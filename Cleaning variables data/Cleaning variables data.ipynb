### Cleaning overview file

```python
import pandas as pd

```

```python

Natgas_overview = pd.read_csv("/Users/richardlesko/Desktop/Data Processing in Python/GitHub - Natural Gas/Python-Final-Project-Natural-gas-price-analysis-/data/raw/EIA_T04_01_Natural_Gas_Overview_Monthly.csv")
Natgas_overview.head()
```

```python
Natgas_overview["year"] = Natgas_overview["YYYYMM"] // 100
Natgas_overview["month"] = Natgas_overview["YYYYMM"] % 100
```

```python
Natgas_overview.head()
```

```python
Natgas_overview["month"] = Natgas_overview["YYYYMM"] % 100
Natgas_overview = Natgas_overview[Natgas_overview["month"] <= 12]
```

```python
Natgas_overview.head()
```

```python
Natgas_overview["year_month"] = Natgas_overview["YYYYMM"]
Natgas_overview.head()
```

```python
Natgas_overview = Natgas_overview.drop(columns=["year", "month", "year_month", "MSN", "Unit"])
Natgas_overview.head()
```

```python
Natgas_overview = Natgas_overview[Natgas_overview["Value"] != "Not Available"]
Natgas_overview.head()
```

```python

Natgas_overview_pivot = Natgas_overview.pivot(index='YYYYMM', columns='Description', values='Value')


desc_order = Natgas_overview[['Description', 'Column_Order']].drop_duplicates().sort_values('Column_Order')
ordered_columns = desc_order['Description'].tolist()


Natgas_overview_pivot = Natgas_overview_pivot[ordered_columns]


Natgas_overview_pivot.reset_index(inplace=True)

print(Natgas_overview_pivot.head())
```

```python
Natgas_overview_pivot.to_csv(
    "/Users/richardlesko/Desktop/Data Processing in Python/GitHub - Natural Gas/Python-Final-Project-Natural-gas-price-analysis-/data/processed/Natgas_overview_monthly_processed.csv",
    index=False
)
```

```python
### converting weekly natgas prices to monthly prices
```

```python


henryhub_weekly_clean = pd.read_csv(
    "/Users/richardlesko/Desktop/Data Processing in Python/GitHub - Natural Gas/Python-Final-Project-Natural-gas-price-analysis-/data/processed/henryhub_weekly_clean.csv"
)

henryhub_weekly_clean.head()
```

```python

daily_cols = ['mon', 'tue', 'wed', 'thu', 'fri']
henryhub_weekly_clean[daily_cols] = henryhub_weekly_clean[daily_cols].apply(pd.to_numeric, errors='coerce')
henryhub_weekly_clean['weekly_avg'] = henryhub_weekly_clean[daily_cols].mean(axis=1, skipna=True)


henryhub_weekly_clean['week_start'] = pd.to_datetime(henryhub_weekly_clean['week_start'])
henryhub_weekly_clean['year'] = henryhub_weekly_clean['week_start'].dt.year
henryhub_weekly_clean['month'] = henryhub_weekly_clean['week_start'].dt.month
henryhub_weekly_clean['year_month'] = henryhub_weekly_clean['week_start'].dt.year * 100 + henryhub_weekly_clean['week_start'].dt.month

henryhub_monthly = henryhub_weekly_clean.groupby('year_month')['weekly_avg'].mean().reset_index()
henryhub_monthly.rename(columns={'weekly_avg': 'monthly_avg_price'}, inplace=True)

print(henryhub_monthly.head())
```

```python
henryhub_monthly.to_csv(
    "/Users/richardlesko/Desktop/Data Processing in Python/GitHub - Natural Gas/Python-Final-Project-Natural-gas-price-analysis-/data/processed/henryhub_monthly.csv",
    index=False
)
```
